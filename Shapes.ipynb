```python
import numpy as np
import plotly.graph_objects as go

#This will get the accuracy of pi
num_points = 10000


x = np.random.uniform(-1, 1, num_points)
y = np.random.uniform(-1, 1, num_points)


distance = np.sqrt(x**2 + y**2)


inside_circle = distance <= 1


pi_estimate = 4 * np.sum(inside_circle) / num_points


fig = go.Figure()


fig.add_trace(go.Scatter(
    x=x[inside_circle],
    y=y[inside_circle],
    mode='markers',
    marker=dict(color='blue'),
    name='Inside Circle'
))

fig.add_trace(go.Scatter(
    x=x[~inside_circle],
    y=y[~inside_circle],
    mode='markers',
    marker=dict(color='red'),
    name='Outside Circle'
))


theta = np.linspace(0, np.pi / 2, 100)
circle_x = np.cos(theta)
circle_y = np.sin(theta)
fig.add_trace(go.Scatter(
    x=np.concatenate([circle_x, [0]]),
    y=np.concatenate([circle_y, [0]]),
    mode='lines',
    line=dict(color='black', width=2),
    name='Circle'
))


fig.update_layout(
    title=f"Estimation of Pi: {pi_estimate:.4f}",
    xaxis=dict(range=[-1, 1]),
    yaxis=dict(range=[-1, 1]),
    showlegend=True
)

fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go


theta = np.linspace(0, 2 * np.pi, 100)


def heart_shape(t, scale):
    x = scale * 16 * (np.sin(theta + t) ** 3)
    y = scale * (13 * np.cos(theta + t) - 5 * np.cos(2 * (theta + t)) - 2 * np.cos(3 * (theta + t)) - np.cos(4 * (theta + t)))
    return x, y


frames = []
scales = np.concatenate([np.linspace(0.5, 1.2, 100), np.linspace(1.2, 0.5, 100)])
for t, scale in zip(np.linspace(0, 2 * np.pi, 200), scales):
    x, y = heart_shape(t, scale)
    frames.append(go.Frame(data=go.Scatter(x=x, y=y, mode='lines', fill='toself', fillcolor='red', line=dict(color='red'))))

x, y = heart_shape(0, 1)
fig = go.Figure(
    data=go.Scatter(x=x, y=y, mode='lines', fill='toself', fillcolor='red', line=dict(color='red')),
    layout=go.Layout(
        title="Popping Heart Shape",
        xaxis=dict(range=[-20, 20], autorange=False),
        yaxis=dict(range=[-20, 20], autorange=False),
        updatemenus=[dict(type='buttons', showactive=False, buttons=[dict(label='Play', method='animate', args=[None, dict(frame=dict(duration=50, redraw=True), fromcurrent=True)])])]
    ),
    frames=frames
)


fig.show()
```

```python
import matplotlib.pyplot as plt

def fibonacci(n):
    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[i-1] + fib[i-2])
    return fib


n = 50


fib_seq = fibonacci(n)

plt.plot(fib_seq, marker='*', linestyle='-')

plt.xlabel('Index')
plt.ylabel('Fibonacci Number')
plt.title('Fibonacci Sequence')

plt.grid(True)
plt.show()
```
